```python
#Take a number from the user which is salary if the salary is more than 50,000 there is 10% tax print his final salary
sal=int(input("Enter your salary :"));
if(sal>=50000):
  tax=sal/10;
  print("Taxes are",tax)
  print("Your salary after taxes is  :",sal-tax)
else:
  print("Your salary is less than 50,000")
  print("No tax reduction for you")
```
